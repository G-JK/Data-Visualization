```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df = pd.read_csv("scatter_data.csv")
```

```python
df.head()
```

```python
# Scatter Plot using Matplotlib
plt.figure(figsize=(12,8))
scatter = plt.scatter(df['Area (ft.)'],
            df['Price'], alpha=0.6,
            c= df['Building Type'],
            cmap = 'viridis')
plt.legend(*scatter.legend_elements(), 
           loc= 'upper left', 
           title= "Building Type")
plt.title("Relation Between Area and Price of California Real Estate", fontsize = 14, weight ='bold')
plt.xlabel("Area (sq ft)", weight = 'bold')
plt.ylabel("Price in (000' $)", weight='bold')
plt.show()
```

```python
# Scatter Plot using Seaborn
plt.figure(figsize=(12,8))
sns.scatterplot( x='Price',
                y='Area (ft.)',
                data= df,
                hue = df['Building Type'], 
                palette=['black','darkblue','purple','pink','white'],
                s= 100)
plt.title("Relation Between Area and Price of California Real Estate", fontsize = 14, weight ='bold')
plt.xlabel("Area (sq ft)", weight = 'bold')
plt.ylabel("Price in (000' $)", weight='bold')
plt.show()
```
