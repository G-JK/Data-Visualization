```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df = pd.read_csv("stacked_area_chart_data.csv")
```

```python
df.head(2)
```

```python
# Stacked Area Chart
colors = ["#011638","#7e2987","#ef2026"]
labels = ["Diesel", "Petrol", "Gas"]
sns.set_style("white")
plt.figure(figsize=(12,6))
plt.stackplot(df["Year"],
              df['Diesel'],
              df['Petrol'],
              df['Gas'],
              colors = colors,
              edgecolors = "none")
plt.xticks(df['Year'], rotation= 45)
plt.legend(labels=labels, loc="upper left")
plt.ylabel("Number of Cars", fontsize=13)
plt.title("Popularity of Fuel engine Types (1982-2016)", fontsize=14, weight="bold")
sns.despine()
plt.show()
```
