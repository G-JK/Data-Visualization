```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df = pd.read_csv("scatter_plot_ii.csv")
```

```python
df
```

```python
# Regression Plot
# plt.figure(figsize=(10,8))
sns.set(rc = {'figure.figsize' :(9,6)})
sns.regplot(x='Budget',
            y = 'Sales',
            data = df,
            scatter_kws = {'color':'k'},
            line_kws = {'color':'red'})
plt.xlabel("Ad Expenditure in (000's $)")
plt.ylabel("Sales in (000's unit)")
plt.title("Effect of Ad Expenditure on Sales", size=16, weight='bold')

plt.show()
```

```python
#LM Plot
sns.lmplot(x= 'Budget',
           y= 'Sales',
           data = df,
           height = 10,
           scatter_kws = {'color':'k'},line_kws = {'color':'red'})
plt.xlabel("Ad Expenditure in (000's $)")
plt.ylabel("Sales in (000's unit)")
plt.title("Effect of Ad Expenditure on Sales", size=16, weight='bold')

plt.show()
```
