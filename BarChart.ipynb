```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
data = pd.read_csv('bar_chart_data.csv')
```

```python
data.head()
```

```python
# Bar Chart
plt.figure(figsize=(9,6))
plt.bar(x=data['Brand'], height=data['Cars Listings'], color = "midnightblue")
plt.xticks(rotation = 45, fontsize = 13)
plt.yticks(fontsize = 13)
plt.title("Car Listing by Brands",fontsize = 16, fontweight = 'bold')
plt.ylabel("Number of Listings", fontsize = 13)
plt.show()
plt.savefig("Used Cars Bar.png")
```
