```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df = pd.read_csv("line_chart_data.csv")
```

```python
df
```

```python
# Data Preprocessing
df["new_date"] = pd.to_datetime(df['Date'])
```

```python
df['new_date']
```

```python
# Line Chart
labels = ["S&P 500", "FTSE 100"]
plt.figure(figsize=(20,8))
plt.plot(df['new_date'],df['GSPC500'])
plt.plot(df['new_date'],df['FTSE100'])
plt.title("S&P vs FTSE Returns (2000 - 2010)", fontsize=14, fontweight='bold')
plt.ylabel("Returns", fontsize=10)
plt.xlabel("Date", fontsize=10)
plt.legend(labels=labels, fontsize='large')
plt.show()
```

```python
df_h2_08 = df[(df.new_date >= '2008-07-01') & (df.new_date <= '2008-12-01') ]
```

```python
df_h2_08
```

```python
# Line Chart
# labels = ["S&P 500", "FTSE 100"]
plt.figure(figsize=(20,8))
plt.plot(df_h2_08['new_date'],df_h2_08['GSPC500'], color='midnightblue')
plt.plot(df_h2_08['new_date'],df_h2_08['FTSE100'], color='crimson')
plt.title("S&P vs FTSE Returns (H2 2008)", fontsize=14, fontweight='bold')
plt.ylabel("Returns", fontsize=10)
plt.xlabel("Date", fontsize=10)
plt.legend(labels=labels, fontsize='large')
plt.show()
```
