```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
data = pd.read_csv("pie_chart_data.csv")
```

```python
data
```

```python
sns.set_palette("colorblind")
```

```python
# Pie Chart
plt.figure(figsize=(10,8))
plt.pie(data['Number of Cars'], labels=data['Engine Fuel Type'].values, 
        autopct='%.2f%%', textprops = {"size":"x-large",
                                       "fontweight":"bold",
                                       "rotation":"30",
                                       "color":"w"})
plt.legend()
plt.title("Cars by Engine Fuel Type", fontsize=18, fontweight="bold")
plt.show()
```
